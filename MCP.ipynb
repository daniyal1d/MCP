```python
detailed_system_prompt = """
You are an expert AI assistant. You must provide:
- Accurate, factual, and up-to-date information.
- Detailed breakdowns (step-by-step if needed).
- Definitions, examples, and comparisons when possible.
- Responses structured with headings, bullet points, or numbered steps.

If you don't know the exact answer, clearly state so. Do not hallucinate or guess.
Be professional and concise, but never vague.
"""

model = genai.GenerativeModel(
    model_name="gemini-2.0-flash",
    system_instruction=detailed_system_prompt # Corrected variable name
)

chat = model.start_chat(history=[])
```

```python
def is_response_vague(text):
    vague_phrases = [
        "I'm not sure", "it depends", "can't say exactly", "might be",
        "could be", "varies", "usually", "generally", "typically",
        "there are many factors", "in some cases", "sometimes"
    ]
    return any(phrase in text.lower() for phrase in vague_phrases)

def improve_response(original_question, current_response):
    improvement_prompt = f"""
You gave the following response:
---
{current_response}
---

This response was too vague or generic. Please rewrite it to be:
- More detailed and specific
- Based on accurate facts
- Including concrete examples or comparisons
- Avoiding uncertainty

Original Question: {original_question}
"""
    fixed = chat.send_message(improvement_prompt)
    return fixed.text
```

```python
def start_accurate_chatbot():
    print("ðŸ“˜ Gemini Chatbot (Max Accuracy Mode) â€” type 'exit' to quit.")

    while True:
        user_input = input("\nðŸ‘¤ You: ")
        if user_input.lower() == 'exit':
            print("\nðŸ”š Chat Ended.")
            break

        response = chat.send_message(user_input)
        reply = response.text.strip()

        # Vague detection & fix
        if is_response_vague(reply):
            print("\nðŸ¤– Gemini: (initial reply too vague, improving...)")
            reply = improve_response(user_input, reply)

        print(f"\nðŸ¤– Gemini:\n{reply}")

        # Optional: Show history
        print("\nðŸ“œ Chat History:")
        for msg in chat.history:
            role = "You" if msg.role == "user" else "Gemini"
            print(f"{role}: {msg.parts[0].text}\n{'-'*60}")
```

```python
start_accurate_chatbot()
```
